#### 1. Load the Data

```python
import pandas as pd

# Load integrated datasets
df_sm = pd.read_csv("Integrated_SweetsMakers.csv")
df_sp = pd.read_csv("Integrated_SweetsPackagers.csv")
df_r = pd.read_csv("Integrated_Retailers.csv")

# Optional: load synthetic orders if needed
df_orders = pd.read_csv("Synthetic_Orders.csv")
```

#### Inspect the Data

```python
# Sweets Makers
print("Sweets Makers Data:")
print(df_sm.head())
print(df_sm.describe())

# Sweets Packagers
print("\nSweets Packagers Data:")
print(df_sp.head())
print(df_sp.describe())

# Retailers
print("\nRetailers Data:")
print(df_r.head())
print(df_r.describe())
```

#### 3. Visualize Key Features

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Plot histograms for Sweets Makers' performance ratios
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_sm["Laddoos_Ratio"], kde=True, color="blue")
plt.title("Distribution of Laddoos Ratio")

plt.subplot(1, 2, 2)
sns.histplot(df_sm["Jalebis_Ratio"], kde=True, color="green")
plt.title("Distribution of Jalebis Ratio")

plt.tight_layout()
plt.show()
```

#### 4. Explore Correlations

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Select only numeric columns
numeric_df_sm = df_sm.select_dtypes(include=[np.number])

# Compute the correlation matrix on numeric columns
corr_matrix_sm = numeric_df_sm.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_sm, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix - Sweets Makers")
plt.show()
```
